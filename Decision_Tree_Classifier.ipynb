```python
# pandas for data handling
import pandas as pd

# train_test_split for splitting dataset into train & test
from sklearn.model_selection import train_test_split

#  TRAIN THE DECISION TREE CLASSIFIER
from sklearn.tree import DecisionTreeClassifier

# mean_squared_error to check model performance
from sklearn.metrics import mean_squared_error

#  CONFUSION MATRIX & CLASSIFICATION REPORT
from sklearn.metrics import confusion_matrix, classification_report

```

```python
# load the dataset
df =pd.read_csv("/content/banknotes.csv")
```

```python
# Display first 5 rows to confirm successful load
df.head()
```

```python
# X - y split
# X contains all independent columns (features)
X = df[["Variance", "Skewness", "Curtosis", "Entropy"]]

# y contains only the output column (target)
y = df["Class"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,     # 20% test data
    random_state=42
)

print("X-y split and train-test split completed!")
```

```python
#  TRAIN THE DECISION TREE CLASSIFIER


model = DecisionTreeClassifier()     # create model
model.fit(X_train, y_train)          # train model
```

```python
#  TEST THE MODEL & GET PREDICTIONS
y_pred = model.predict(X_test)
```

```python
#  CONFUSION MATRIX & CLASSIFICATION REPORT
from sklearn.metrics import confusion_matrix, classification_report

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
#  PREDICT A NEW SAMPLE
new_sample = [[3.0, 7.0, -2.0, -1.0]]

prediction = model.predict(new_sample)

print("Prediction for new sample:", prediction[0])
```
